# Question: Prisoners Boat Problem 
##### • 3 Guards, 3 Prisoners Needs To Cross A River.
##### • Boat Take One Or Two At A Time.
##### • Prisoners May Never Outnumber Guards On Either Coast, Thought Prisoners May Be Alone On Either Coast.
### Write Python Code For Three Prisoners And Three Guards Problem Some Picture To Explain


# Solved by breadth first search (BFS) algorithm:

```python
import numpy as np

class State:
    def __init__(self, rg, rp, lg=0, lp=0, boat_pos='left', level=0):
        self.right_gurads = rg
        self.right_prisoners = rp
        self.left_gurads = lg
        self.left_prisoners = lp
        self.boat_pos = boat_pos
        self.level = level
        self.state = [np.array([rg,rp,lg,lp]),boat_pos,level]

    def check_state(self):
        if self.right_gurads == 0 and self.right_prisoners == 0:
            return "final state"
        elif ( (self.right_prisoners <= self.right_gurads) or (self.right_gurads == 0) ) and ( (self.left_prisoners <= self.left_gurads) or (self.left_gurads == 0) ):
            return True
        else:
            return False
        
    def get_list_state(self):
        return self.state
    
            
        
class Boat:
    def __init__(self, State, load=2):
        self.load = load
        self.boat_pos = State.boat_pos
        self.state = State.state[0]
        self.comps = self.get_compositions()
        self.level = State.level
        
    def switch_boat(self):
        if self.boat_pos == "left":
            self.boat_pos = "right"
        elif self.boat_pos == "right":
            self.boat_pos = "left"
        
    def get_compositions(self):
        self.comps = []
        for i in range(self.load+1):
            for j in range(self.load+1):
                if (i+j ==0) or (i+j > self.load):
                    continue
                else :
                    self.comps.append(np.array([i,j]))
        return self.comps
    
    def get_next_states(self):
        next_states = []
        for comp in self.comps:
            if self.boat_pos == "left":
                state = self.state + np.append(-comp , comp)
                if sum(state < 0) > 0:                        #remove negative results 
                    continue 
                rg,rp,lg,lp = state
                state = State(rg,rp,lg,lp, boat_pos="right", level = self.level + 1)
                next_states.append(state)
            elif self.boat_pos == "right":
                state = self.state + np.append(comp , -comp)
                if sum(state < 0) > 0:                        #remove negative results 
                    continue
                rg,rp,lg,lp = state
                state = State(rg,rp,lg,lp , boat_pos="left", level = self.level + 1 )
                next_states.append(state)
        self.switch_boat()
        return next_states
    
                
    def show_compositions(self):
        print(f'boat can transfer {self.comps} "(guards,prisoners)"')
        


def boat_problem():
    guards_no = int(input("Enter number of guards :"))
    prisoners_no = int(input("Enter number of prisoners :"))
    boat_load = int(input("Enter boat load :"))
        
    init = State(guards_no,prisoners_no)
    processing_paths = [[init]]
    checked_paths = []
    stop = 0
    while stop != True:
        for j in range(len(processing_paths)):
            current = processing_paths.pop(0)
            a= [ [list(item.state[0]),item.state[1]] for item in current]
            if current[-1].right_gurads == 0 and current[-1].right_prisoners == 0:
                a.append('final state')
                checked_paths.append(a)
                stop = True
            elif (a[-1] in a[:-1]) :
                a.append('Endless loop')
                checked_paths.append(a)
            elif current[-1].check_state() == False :
                a.append('Failed')
                checked_paths.append(a)
            elif current[-1].check_state() == True :
                boat = Boat(current[-1], load = boat_load )
                for s in boat.get_next_states():
                    temp_current = current.copy()
                    temp_current.append(s)
                    processing_paths.append(temp_current)
              
            
    
    #checking remaining paths
    #print(processing_paths)
    #for i in processing_paths:
    #    for j in i :
    #        print(j.state)
    final_paths = []
    for i in checked_paths:
        if i[-1] == "final state":
            final_paths.append(i)
    
    print(f'you need to move the boat {len(final_paths[0])-2} steps at minimum to transfer {prisoners_no} prisoners and {guards_no} guards to the other side with a boat can carry up to {boat_load} persons in {len(final_paths)} ways')
    show = int(input("Enter 0 to show the full paths : \n"))
    if show == False:
        for path in final_paths:
            print(path)
            print("*************************************************")
    
    
boat_problem()
```
